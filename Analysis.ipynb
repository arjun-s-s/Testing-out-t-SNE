```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.manifold import TSNE
from sklearn import datasets
from sklearn.datasets import fetch_openml

```

```python
mnist = datasets.fetch_openml("mnist_784")
X, y = mnist.data / 255.0, mnist.target.astype(int)


# Create a 2x5 grid for displaying the images
fig, axs = plt.subplots(2, 5, figsize=(12, 6))

for i in range(10):
    row, col = divmod(i, 5)  # Calculate row and column indices
    axs[row, col].imshow(X.iloc[i].values.reshape(28, 28), cmap="gray")
    axs[row, col].axis("off")  # Hide axes
    axs[row, col].set_title(f"Image {i + 1}")

plt.tight_layout()  # Adjust spacing between subplots
plt.show()

```

```python
# Load the MNIST dataset
mnist = datasets.fetch_openml("mnist_784")
X, y = mnist.data / 255.0, mnist.target.astype(int)

# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Visualize the t-SNE results
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap="jet", marker=".")
plt.colorbar(scatter)
plt.title("t-SNE Visualisation of MNIST Dataset")
plt.show()
```

```python
# Load the MNIST dataset
mnist = fetch_openml("mnist_784")
X, y = mnist.data / 255.0, mnist.target.astype(int)

# Apply t-SNE with 3 components
tsne = TSNE(n_components=3, random_state=42)
X_tsne = tsne.fit_transform(X)

# Visualise the t-SNE results in 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection="3d")
scatter = ax.scatter(X_tsne[:, 0], X_tsne[:, 1], X_tsne[:, 2], c=y, cmap="jet", marker=".")
plt.colorbar(scatter)
plt.title("t-SNE Visualisation of MNIST Dataset (3D)")
plt.show()
```
